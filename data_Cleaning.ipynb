```python
# import the required python liraries
import pandas as pd
import numpy as np

# Read the dataset  
df = pd.read_csv('Harry Potter All movie Chr Dialogs.csv')
df.head()
```

```python
# In our dataset we have some unwanted columns like (movie,chapter,character,dialog).
# we can only using the character and dialog columns. For that we have to remove movie and chapter columns.
df1 = df.drop(['movie'],axis=1)
df2 = df1.drop(['chapter'],axis=1)
df2.head()
```

```python
# Now we are check that any null value are there in datsset.
df2.isnull()
```

```python
# By using the sum() method it is return the number of null value present in each columns 
df2.isnull().sum()
```

```python
# Find the row number that have null values present
df3 = pd.isnull(df2["dialog"])
df2[df3]
```

```python
# remove the row that have null value
df2 = df2.drop(1421)
```

```python
# Now check any null value present or not 
df2.isnull().sum()
```

```python
# In character column we have only Harry Potter dialogs. So we can remove all other dialogs.
df3 =  df2[df2['character'] == 'Harry Potter']
df3
```

```python
# Finally our data is clean so we can save that clean data in new csv file. 
df3.to_csv('harry_rows.csv', index=False)
```

```python
#Merge to columns 
df3['newcol'] = df2['character'] +" : " +df2['dialog']
df3
```

```python
# After merge two column remove the character and dialog columns
df4 = df3.drop(['character','dialog'],axis='columns')
df4
```

```python
# Save to the new csv file
df3.to_csv('harry_rows_final.csv', index=False)
```
